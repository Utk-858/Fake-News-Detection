```python
import feedparser
import pandas as pd
import time
import sqlite3
from bs4 import BeautifulSoup
from tenacity import retry, stop_after_attempt, wait_fixed
from datetime import datetime

# Define RSS feeds with source types and names
rss_feeds = [
    # Credible Sources
    {"url": "http://feeds.bbci.co.uk/news/rss.xml", "source_type": "credible", "name": "BBC News"},
    {"url": "http://feeds.reuters.com/reuters/topNews", "source_type": "credible", "name": "Reuters"},
    {"url": "https://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml", "source_type": "credible", "name": "The New York Times"},
    {"url": "https://www.npr.org/rss/rss.php?id=1001", "source_type": "credible", "name": "NPR News"},
    {"url": "https://feeds.nbcnews.com/nbcnews/public/news", "source_type": "credible", "name": "NBC News"},
    {"url": "https://www.aljazeera.com/xml/rss/all.xml", "source_type": "credible", "name": "Al Jazeera"},
    {"url": "https://www.dw.com/en/rss", "source_type": "credible", "name": "Deutsche Welle"},

    # Biased Sources
    {"url": "http://feeds.foxnews.com/foxnews/latest", "source_type": "biased", "name": "Fox News"},
    {"url": "https://www.theguardian.com/world/rss", "source_type": "biased", "name": "The Guardian"},
    {"url": "http://rss.cnn.com/rss/cnn_topstories.rss", "source_type": "biased", "name": "CNN"},
    {"url": "https://www.msnbc.com/msnbc/rss", "source_type": "biased", "name": "MSNBC"},

    # Questionable Sources
    {"url": "https://www.breitbart.com/feed/", "source_type": "questionable", "name": "Breitbart"},
    {"url": "https://www.infowars.com/rss.xml", "source_type": "questionable", "name": "InfoWars"},

    # Fact-Checking
    {"url": "https://www.snopes.com/feed/", "source_type": "fact-checking", "name": "Snopes"},
    {"url": "https://www.politifact.com/feeds/articles/", "source_type": "fact-checking", "name": "PolitiFact"},

    # Aggregators
    {"url": "https://news.google.com/rss/search?q=technology&hl=en-US&gl=US&ceid=US:en", "source_type": "aggregator", "name": "Google News Technology"},
    {"url": "https://news.google.com/rss/search?q=world&hl=en-US&gl=US&ceid=US:en", "source_type": "aggregator", "name": "Google News World"},
    {"url": "https://www.ap.org/rss", "source_type": "credible", "name": "Associated Press"},
    {"url": "https://feeds.a.dj.com/rss/RSSWorldNews.xml", "source_type": "credible", "name": "The Wall Street Journal"},

]

# Retry decorator for fetching feeds
@retry(stop=stop_after_attempt(3), wait=wait_fixed(2))
def fetch_feed(url):
    return feedparser.parse(url)

# Clean text (remove HTML tags and extra whitespace)
def clean_text(text):
    if text:
        soup = BeautifulSoup(text, "html.parser")
        return ' '.join(soup.get_text().strip().split())
    return ""

# Initialize SQLite database
def init_db():
    conn = sqlite3.connect("news_articles.db")
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS articles (
            title TEXT,
            link TEXT,
            published TEXT,
            source TEXT,
            description TEXT,
            source_type TEXT,
            label TEXT,
            scraped_at TEXT
        )
    """)
    conn.commit()
    return conn

try:
    # Initialize database
    conn = init_db()

    # List to store articles
    articles = []

    # Process each feed
    for feed_info in rss_feeds:
        rss_url = feed_info["url"]
        source_type = feed_info["source_type"]
        source_name = feed_info["name"]

        print(f"Fetching feed: {source_name} ({rss_url})")
        feed = fetch_feed(rss_url)

        if feed.bozo:
            print(f"Error fetching feed {source_name}: {feed.bozo_exception}")
            continue

        # Extract data
        for entry in feed.entries:
            article = {
                "title": clean_text(entry.get("title", "")),
                "link": entry.get("link", ""),
                "published": entry.get("published", ""),
                "source": source_name,
                "description": clean_text(entry.get("description", "") or entry.get("summary", "")),
                "source_type": source_type,
                "label": "unknown",  # Placeholder for labeling
                "scraped_at": datetime.now().isoformat()
            }
            articles.append(article)

        # Respect rate limits
        time.sleep(3)  # Increased delay for politeness

    # Convert to DataFrame
    df = pd.DataFrame(articles)

    # Save to SQLite
    df.to_sql("articles", conn, if_exists="append", index=False)

    # Save to CSV for backup
    df.to_csv("news_articles.csv", index=False, encoding="utf-8")
    print(f"Collected {len(articles)} articles. Saved to news_articles.db and news_articles.csv")

    # Print first 5 articles
    for article in articles[:5]:
        print(f"Title: {article['title']}")
        print(f"Link: {article['link']}")
        print(f"Published: {article['published']}")
        print(f"Source: {article['source']}")
        print(f"Description: {article['description'][:100]}...")
        print(f"Source Type: {article['source_type']}")
        print(f"Label: {article['label']}")
        print(f"Scraped At: {article['scraped_at']}")
        print("-" * 50)

except Exception as e:
    print(f"Failed to process feeds: {e}")

finally:
    # Close database connection
    conn.close()
```

```python
import pandas as pd
import sqlite3
import requests
from sentence_transformers import SentenceTransformer, util
from bs4 import BeautifulSoup
import time
from tenacity import retry, stop_after_attempt, wait_fixed
from datetime import datetime
from dotenv import load_dotenv
import os

# Google Fact Check Tools API key
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")  # Replace with your actual API key

# Load dataset from SQLite
def load_dataset():
    conn = sqlite3.connect("news_articles.db")
    df = pd.read_sql("SELECT * FROM articles", conn)
    conn.close()
    return df

# Save updated dataset
def save_dataset(df):
    conn = sqlite3.connect("news_articles.db")
    df.to_sql("articles", conn, if_exists="replace", index=False)
    df.to_csv("news_articles_labeled.csv", index=False, encoding="utf-8")
    conn.close()

# Clean text (remove HTML tags and normalize whitespace)
def clean_text(text):
    if text:
        soup = BeautifulSoup(text, "html.parser")
        return ' '.join(soup.get_text().strip().split())
    return ""

# Retry decorator for API requests
@retry(stop=stop_after_attempt(3), wait=wait_fixed(2))
def fetch_fact_checks(query, api_key):
    url = f"https://factchecktools.googleapis.com/v1alpha1/claims:search?query={query}&key={api_key}"
    response = requests.get(url)
    response.raise_for_status()
    return response.json().get("claims", [])

try:
    # Load dataset
    df = load_dataset()
    print(f"Loaded {len(df)} articles from news_articles.db")

    # Initialize fact-checks list
    fact_checks = []

    # Fetch fact-checks for each unique source (to avoid excessive API calls)
    unique_sources = df["source"].unique()
    for source in unique_sources:
        print(f"Fetching fact-checks for source: {source}")
        claims = fetch_fact_checks(source, GOOGLE_API_KEY)

        for claim in claims:
            review = claim.get("claimReview", [{}])[0]
            verdict = review.get("textualRating", "").lower()
            fact_checks.append({
                "title": clean_text(claim.get("text", "")),
                "description": clean_text(review.get("title", "")),
                "link": review.get("url", ""),
                "source": review.get("publisher", {}).get("name", "Unknown"),
                "verdict": verdict
            })

        time.sleep(1)  # Respect API rate limits

    # Convert fact-checks to DataFrame
    fact_df = pd.DataFrame(fact_checks)
    print(f"Collected {len(fact_checks)} fact-checks from Google Fact Check Tools")

    # Save fact-checks for reference
    fact_df.to_csv("fact_checks.csv", index=False, encoding="utf-8")

    # Load sentence transformer model
    model = SentenceTransformer("all-MiniLM-L6-v2")

    # Combine title and description for similarity
    article_texts = (df["title"] + " " + df["description"]).fillna("").tolist()
    fact_texts = (fact_df["title"] + " " + fact_df["description"]).fillna("").tolist()

    # Encode texts
    article_embeddings = model.encode(article_texts, convert_to_tensor=True, show_progress_bar=True)
    fact_embeddings = model.encode(fact_texts, convert_to_tensor=True, show_progress_bar=True) if fact_texts else None

    # Initialize new columns
    df["fact_check_link"] = ""
    df["fact_check_verdict"] = ""
    df["label"] = df.get("label", "unknown")

    # Assign labels based on similarity and verdict
    if fact_texts and fact_embeddings is not None:
        cosine_scores = util.cos_sim(article_embeddings, fact_embeddings)
        similarity_threshold = 0.3  # Adjust based on validation

        for i, row in df.iterrows():
            max_score = cosine_scores[i].max().item()
            if max_score > similarity_threshold:
                fact_idx = cosine_scores[i].argmax().item()
                fact = fact_df.iloc[fact_idx]
                df.at[i, "fact_check_link"] = fact["link"]
                df.at[i, "fact_check_verdict"] = fact["verdict"]

                # Map verdict to label
                verdict = fact["verdict"].lower()
                if "true" in verdict or "mostly true" in verdict:
                    df.at[i, "label"] = "real"
                elif "false" in verdict or "mostly false" in verdict or "pants on fire" in verdict:
                    df.at[i, "label"] = "fake"
                elif "half true" in verdict or "partly" in verdict or "mixture" in verdict:
                    df.at[i, "label"] = "partially fake"
                else:
                    df.at[i, "label"] = "unknown"

    # Save updated dataset
    save_dataset(df)

    # Print summary
    print("Fact-checking labeling completed. Label distribution:")
    print(df["label"].value_counts())
    print("\nSample of matched articles:")
    matched = df[df["fact_check_link"] != ""][["title", "source", "label", "fact_check_verdict", "fact_check_link"]]
    print(matched.head().to_string(index=False))

    # Save sample for manual review
    matched.sample(min(10, len(matched))).to_csv("matched_articles_sample.csv", index=False)
    print("Saved sample of matched articles to matched_articles_sample.csv for review")

except Exception as e:
    print(f"Error during fact-checking labeling: {e}")

finally:
    # Ensure database connection is closed
    if 'conn' in locals():
        conn.close()
```

```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset from SQLite or CSV
def load_dataset():
    try:
        conn = sqlite3.connect("news_articles.db")
        df = pd.read_sql("SELECT * FROM articles", conn)
        conn.close()
        print(f"Loaded {len(df)} articles from news_articles.db")
        return df
    except Exception as e:
        print(f"Error loading from database: {e}. Trying CSV...")
        try:
            df = pd.read_csv("news_articles_labeled.csv")
            print(f"Loaded {len(df)} articles from news_articles_labeled.csv")
            return df
        except Exception as e:
            print(f"Error loading CSV: {e}")
            return None

# Save updated dataset
def save_dataset(df):
    try:
        conn = sqlite3.connect("news_articles.db")
        df.to_sql("articles", conn, if_exists="replace", index=False)
        conn.close()
        df.to_csv("news_articles_labeled.csv", index=False, encoding="utf-8")
        print("Saved updated dataset to news_articles.db and news_articles_labeled.csv")
    except Exception as e:
        print(f"Error saving dataset: {e}")

try:
    # Load dataset
    df = load_dataset()
    if df is None:
        raise Exception("Failed to load dataset")

    # Check if labels exist
    valid_labels = ["real", "fake", "partially fake"]
    label_counts = df[df["label"].isin(valid_labels)]["label"].value_counts()

    # Apply source-based labeling if no valid labels
    if label_counts.empty or len(label_counts) < 2:
        print("No or insufficient 'real', 'fake', or 'partially fake' labels. Applying source-based labeling...")
        source_reliability = {
            "BBC News": "real", "Reuters": "real", "The New York Times": "real",
            "NPR News": "real", "NBC News": "real", "Al Jazeera": "real",
            "Deutsche Welle": "real", "The Guardian": "partially fake",
            "CNN": "partially fake", "Fox News": "partially fake",
            "MSNBC": "partially fake", "Breitbart": "fake", "InfoWars": "fake",
            "Snopes": "unknown", "PolitiFact": "unknown",
            "Google News Technology": "unknown", "Google News World": "unknown"
        }
        df["label"] = df["source"].map(source_reliability).fillna("unknown")
        # Save updated dataset
        save_dataset(df)
        label_counts = df[df["label"].isin(valid_labels)]["label"].value_counts()

    # Print label distribution
    print("Label distribution:")
    print(label_counts)

    # Check if there are enough labels to plot
    if label_counts.empty:
        raise Exception("No articles with 'real', 'fake', or 'partially fake' labels to plot")

    # Create bar graph with seaborn
    plt.figure(figsize=(8, 6))
    sns.barplot(x=label_counts.index, y=label_counts.values, palette=["green", "red", "orange"])
    plt.title("Distribution of Real vs. Fake News Articles")
    plt.xlabel("Label")
    plt.ylabel("Number of Articles")
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Add count labels on top of bars
    for i, count in enumerate(label_counts):
        plt.text(i, count + 0.5, str(count), ha="center", va="bottom")

    plt.tight_layout()

    # Save and show plot
    plt.savefig("real_vs_fake_bar_graph.png")
    plt.show()

    # Save sample of labeled articles for validation
    sample = df[df["label"].isin(valid_labels)][["title", "source", "label"]].sample(min(10, len(df)), random_state=42)
    sample.to_csv("labeled_articles_sample.csv", index=False)
    print("Saved sample of labeled articles to labeled_articles_sample.csv for validation")

except Exception as e:
    print(f"Error creating bar graph: {e}")
```
